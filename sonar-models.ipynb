Add PyPlot for plots and bring in sonar models

```julia
using PyPlot;
include("probmodel.jl");
include("sonarmmodel.jl");
```

Create occupancy vector (assume very high occupancy probability in bins 90-95) 

```julia
N = 100
p_o = 0.005*ones(N,1);
p_o[90:95] = 0.9;
```

Set the probabilities for true positives and true negatives

```julia
p_tp = 0.8; # true positive ( = 1 - false negative)
p_tn = 0.9;  # true negative ( = 1 - false positive)
```

Run first model and check if the sum of the probability distribution function is 1

```julia
pr_a = sonarpmodel(p_o, p_tp, p_tn);
```

Do the same for the second model.

```julia
pr_b = sonarmmodel(p_o, p_tp);
```

Plot results

```julia
plot(1:N,p_o,label="occupancy")
plot(1:N,pr_a,label="TP/TN");
plot(1:N,pr_b,label="TP only (Moravec)");
xlabel("range bin")
legend(loc="upper left")
```
